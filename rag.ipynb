````python
import os
import warnings

warnings.filterwarnings("ignore")
````

````python
model_name = "BAAI/bge-m3"
model_kwargs = {'device': 'cuda'}
encode_kwargs = {'normalize_embeddings': True}
model = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
````

````python
# 벡터 저장소를 저장할 디렉토리
save_directory = "./Chroma"

# 벡터저장소가 이미 존재하는지 확인
if os.path.exists(save_directory):
    shutil.rmtree(save_directory)
    print(f"Directory {save_directory} has been deleted...\n")
````

````python
print("Initiate document vectorization")
chroma_retriever = Chroma.from_documents(chunks, model, persist_directory=save_directory).as_retriever()
bm25_retriever = BM25Retriever.from_documents(
    chunks
)
print('The document database has been created')
````

````python
import json

def save_embeddings_to_file(retriever, file_path='embeddings_output.json'):
    # 벡터스토어에서 모든 문서, 임베딩, 메타데이터 가져오기
    documents = retriever.vectorstore.get()

    # 저장할 데이터 구조화
    data = []
    for i, doc in enumerate(documents['documents']):
        # 각 문서의 임베딩 계산
        embedding = model.embed_documents([doc])[0]
        metadata = documents['metadatas'][i] if documents['metadatas'] else None

        data.append({
            "document": doc,
            "embedding": embedding,
            "metadata": metadata
        })

    # JSON 파일로 저장
    with open(file_path, 'w') as f:
        json.dump(data, f, indent=4)

    print(f"Embedding result saved in '{file_path}'")
````

````python
save_embeddings_to_file(chroma_retriever, file_path='output/embeddings_output.json')
````

````python
ensemble_retriever = EnsembleRetriever(
    retrievers=[bm25_retriever, chroma_retriever], 
    weights=[0.5, 0.5], 
    search_type="similarity_score_threshold",
)
````

````python
from langchain_openai import ChatOpenAI
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
from langchain.prompts import ChatPromptTemplate
from langchain.schema.runnable import RunnableLambda, RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
````

````python
llm = ChatOpenAI(
    model_name="gpt-4o",
    api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
)
````

````python
from langchain.prompts import ChatPromptTemplate

# 질문에 대한 답변을 찾기 위한 프롬프트
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """
            You are a language model designed to analyze and interpret structured data embedded within specific tags in a text format.
            Your primary task is to extract structured information from the input text based on predefined attributes and nested structures. 
            Each attribute and value pair is represented within specific symbols, with the main attributes and their nested sub-attributes appearing within unique tags.

            **Important**: Always respond in Korean.
            
            **Role:**
            1. Identify main attributes and nested sub-attributes in the given text using the tags specified.
            2. Extract data between `[표 속성]` and `[표 내용]` markers. Values are enclosed between triple backticks (```).
            3. Parse main attributes and their sub-attributes, separating them using `' / '`. The sub-attributes are further nested within `{{}}` and `[]`.

            **Instructions:**
            - Carefully analyze the structure within `{{}}` and `[]` to determine levels of nested sub-attributes.
            - For each main attribute, create a hierarchical view of sub-attributes based on their nested positions.
            - Ensure that any extracted structure is clear and easily understandable, reflecting the nested relationships within the attribute values.
            - If additional context is provided, apply it to accurately interpret any ambiguous nested structures.

            Based on these guidelines, analyze the given text and extract attributes and sub-attributes as outlined above, returning the information in a structured format.

            **Context:** {context}
            """
        ),
        (
            "human",
            """
            **Input Text**                
            ```text
            {question}
            ```               
            """
        ),
    ]
)
````

````python
def format_docs(docs):
    # 검색한 문서 결과를 하나의 문단으로 합쳐줍니다.
    return "\n\n".join(doc.page_content for doc in docs)
````

````python
# 단계 8: 체인 생성(Create Chain)
chain = (
    {"context": ensemble_retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
````

````python
with open('./input/question.txt', 'r') as file:
    querys = [line.strip() for line in file]
````

````python
response = [chain.invoke(query) for query in querys]
````

````python
directory = './output'
file_name = 'response.txt'

if not os.path.exists(directory):
    os.makedirs(directory)
    
file_path = os.path.join(directory, file_name)
````

````python
with open(file_path, 'w') as file:
    for query, res in zip(querys, response):
        file.write(f"Query: {query}\nResponse:\n{res}\n===========================================\n")

print(f"Saved in '{file_path}'")
````
